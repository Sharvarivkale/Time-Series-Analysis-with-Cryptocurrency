```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load dataset
df = pd.read_csv("binance_crypto_data_file (1).csv")

# Display first rows
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
# Convert timestamp (DD-MM-YYYY format)
df['timestamp'] = pd.to_datetime(df['timestamp'], dayfirst=True)

# Sort and set index
df = df.sort_values('timestamp')
df.set_index('timestamp', inplace=True)

df.head()
```

```python
df.rename(columns={
    'open': 'Open',
    'high': 'High',
    'low': 'Low',
    'close': 'Close',
    'volume': 'Volume'
}, inplace=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df['Price_Change'] = df['Close'] - df['Open']
df['High_Low_Diff'] = df['High'] - df['Low']
df['Daily_Return'] = df['Close'].pct_change()
```

```python
df['MA_7'] = df['Close'].rolling(7).mean()
df['MA_14'] = df['Close'].rolling(14).mean()
df['MA_30'] = df['Close'].rolling(30).mean()
```

```python
df['Volatility_7'] = df['Daily_Return'].rolling(7).std()
df['Volatility_14'] = df['Daily_Return'].rolling(14).std()
```

```python
df['Momentum_7'] = df['Close'] - df['Close'].shift(7)
df['Momentum_14'] = df['Close'] - df['Close'].shift(14)
```

```python
df['Volume_Change'] = df['Volume'].pct_change()
df['Volume_MA_7'] = df['Volume'].rolling(7).mean()
```

```python
df['Lag_1'] = df['Close'].shift(1)
df['Lag_3'] = df['Close'].shift(3)
df['Lag_7'] = df['Close'].shift(7)
```

```python
df.dropna(inplace=True)
df.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA_7'], label='MA 7')
plt.plot(df['MA_30'], label='MA 30')
plt.legend()
plt.title("Cryptocurrency Price with Moving Averages")
plt.show()
```

```python
features = [
    'Open', 'High', 'Low', 'Volume',
    'Price_Change', 'High_Low_Diff',
    'Daily_Return',
    'MA_7', 'MA_14', 'MA_30',
    'Volatility_7',
    'Momentum_7',
    'Volume_MA_7',
    'Lag_1', 'Lag_3', 'Lag_7'
]

target = 'Close'

X = df[features]
y = df[target]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
train_size = int(len(X_scaled) * 0.8)

X_train = X_scaled[:train_size]
X_test = X_scaled[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]
```

```python
for i in range(1, 30):
    df[f'lag_{i}'] = df['Close'].shift(i)
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X_train)
y_scaled = scaler_y.fit_transform(y_train.values.reshape(-1,1))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))

print("MAE:", mae)
print("RMSE:", rmse)
```
